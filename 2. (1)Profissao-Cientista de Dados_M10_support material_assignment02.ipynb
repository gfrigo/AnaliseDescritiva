````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

# Módulo 10 tarefa 2

Considere a mesma base de dados utilizada na tarefa 1 ```previsao_de_renda.csv```.

````python
filepath = 'https://raw.githubusercontent.com/rhatiro/Curso_EBAC-Profissao_Cientista_de_Dados/main/Mo%CC%81dulo%2010%20-%20Descritiva%20II%20-%20Ana%CC%81lise%20descritiva%20para%20resposta%20conti%CC%81nua/database/Profissa%CC%83o-%20Cientista%20de%20Dados_M10_support%20material.csv'
df = pd.read_csv(filepath_or_buffer=filepath)
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
````

#### 1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
fig, axs = plt.subplots(1, 2, figsize=(10, 10))

ax1 = sns.pointplot(x="posse_de_imovel", y="renda", hue = 'sexo',
                   data=df, dodge=True, errorbar=('ci', 95), ax=axs[0])
plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
ax1.set_title('Renda por Posse de Imóvel')
ax1.set_xlabel('Posse de Imóvel')
ax1.set_ylabel('Renda')

ax2 = sns.pointplot(x="posse_de_veiculo", y="renda", hue = 'sexo',
                   data=df, dodge=True, errorbar=('ci', 95), ax=axs[1])
plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
ax2.set_title('Renda por Posse de Veículo')
ax2.set_xlabel('Posse de Veículo')
ax2.set_ylabel('Renda')

plt.tight_layout()
````

* Quem tem imóveis, geralmente tem uma renda mais alta;
* A renda de homens sem veículo é menor, porém, pouco explicativa, já que em amabos os gráficos a renda de homens é maior. Já para as mulheres ter imóvel e ter veículo justifica renda maiores.

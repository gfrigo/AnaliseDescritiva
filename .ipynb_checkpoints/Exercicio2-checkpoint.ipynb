```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='whitegrid')
plt.rc('figure', figsize=(5,5))
```

```python
df = pd.read_csv('SINASC_RO_2019.csv')
df.head(2)
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
df['IDADEMAE'].head(2)
```

```python
df['IDADEMAE'].isna().sum()
```

```python
fig,axes = plt.subplots(1,1)
sns.boxplot(y='IDADEMAE', data=df)
```

####  2) Faça um boxplot do peso do bebê

```python
df['PESO'].head()
```

```python
df['PESO'].isna().sum()
```

```python
plt.close('All')
fig,axes = plt.subplots(2,2, figsize=(8,8))
sns.boxplot(ax=axes[0,0], y='PESO', data=df)
axes[0,0].set_title('Pesos Gerais')
sns.boxplot(ax=axes[0,1], y='PESO', data=df[df['PESO']>3000])
axes[0,1].set_title('Peso Acima de 3000')
sns.boxplot(ax=axes[1,0],y='PESO', data=df[df['PESO']<3000])
axes[1,0].set_title('Peso Abaixo de 3000g')
plt.tight_layout()
```

```python
df['PESO'].describe()
```

####  3) Faça um histograma da idade da mãe

```python
plt.close('All')
sns.displot(x='IDADEMAE',
            data=df,
            alpha=.5,
            bins=60,
            element='step',
            kde=True
           )
```

#### 4) Faça um histograma do peso do bebê

```python
plt.close('All')
sns.displot(
    x='PESO',
    data=df,
    alpha=.5,
    element='step',
    kde=True
)
```

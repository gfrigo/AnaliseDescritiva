#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df['APGAR5'].isna().sum()
```

```python
df = df.dropna(subset=['APGAR5'])
```

```python
df['APGAR5'].isna().sum()
```

```python
df.loc[(df['APGAR5'] >=0) & (df['APGAR5']<=3), 'APGAR5_CAT'] = 'asfixia_grave'
df.loc[(df['APGAR5'] >=4) & (df['APGAR5']<=5), 'APGAR5_CAT'] = 'asfixia_moderada'
df.loc[(df['APGAR5'] >=6) & (df['APGAR5']<=7), 'APGAR5_CAT'] = 'asfixia_leve'
df.loc[(df['APGAR5'] >=8) & (df['APGAR5']<=10), 'APGAR5_CAT'] = 'asfixia_normal'
```

```python
df.head(2)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
freq = pd.crosstab(df['GESTACAO'], df['APGAR5'])
freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq_pct = freq.div(freq.sum(axis=1), axis=0)
freq_pct
```

```python
freq_pct.plot.bar()
```
